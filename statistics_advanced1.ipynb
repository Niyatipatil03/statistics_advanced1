Question 2: What are the types of random variables?


*  Discrete random variables: Take countable distinct values (e.g., number of heads in 10 coin flips, values of a die). They are described by probability mass functions (PMFs).

*  Continuous random variables: Take values in an interval or continuum (e.g., height, weight, measurement errors). They are described by probability density functions (PDFs).

Question 7: What is the significance of confidence intervals in statistical analysis?
* A confidence interval (CI) gives a range of plausible values for an unknown population parameter (e.g., the mean) based on sample data and a chosen confidence level (e.g., 95%). Interpretation: a 95% CI means that, under repeated sampling and applying the same CI method, about 95% of such intervals would contain the true parameter. CIs quantify estimation uncertainty and are more informative than a single-point estimate because they show precision and reliability.

Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=1000)
mean_value = np.mean(data)
std_value = np.std(data, ddof=1)
print("Mean of generated data:", mean_value)
print("Standard Deviation of generated data:", std_value)
plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.

‚óè Write the Python code to compute the mean sales and its confidence interval.

* The Central Limit Theorem (CLT) states that when we take repeated samples of size n from any population with mean (Œº) and standard deviation (œÉ), the sampling distribution of the sample mean (xÃÑ) tends to follow a normal distribution as n becomes large, regardless of the shape of the population distribution.
* In this case:

  * The daily_sales data represents a sample from the population of all possible daily sales.

  * We can use the sample mean (xÃÑ) and sample standard deviation (s) to estimate the population mean (Œº).

  * The 95% confidence interval (CI) gives a range within which the true population mean is likely to fall.

  * The formula for a 95% confidence interval of the mean is:

    image.png


```python
import numpy as np
from scipy import stats

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to numpy array
sales = np.array(daily_sales)

# Calculate mean, standard deviation, and sample size
mean_sales = np.mean(sales)
std_sales = np.std(sales, ddof=1)
n = len(sales)

# Calculate standard error
se = std_sales / np.sqrt(n)

# Calculate t-critical value for 95% confidence level
confidence = 0.95
t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)

# Calculate confidence interval
margin_of_error = t_crit * se
lower_limit = mean_sales - margin_of_error
upper_limit = mean_sales + margin_of_error

# Display results
print("Mean Sales:", mean_sales)
print("Standard Deviation:", std_sales)
print("Standard Error:", se)
print("t-critical value:", t_crit)
print(f"95% Confidence Interval: ({lower_limit:.2f}, {upper_limit:.2f})")
```
